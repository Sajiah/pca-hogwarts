```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from matplotlib.pyplot import imread
import numpy as np
```

```python
img = imread('hogwarts.jpg')
img = img.astype(np.uint8)
img = img/255
img = img.mean(axis=2)
plt.imshow(img, cmap = "gray")
```

```python
# takes a percentage between 0 and 100
def transform(percentage):
    percentage = percentage / 100
    hogwarts_pca = PCA(n_components=percentage).fit(img)
    transformed = hogwarts_pca.transform(img)
    projection = hogwarts_pca.inverse_transform(transformed)
    plt.imshow(projection, cmap = "gray")
```

```python
transform(95)
```
